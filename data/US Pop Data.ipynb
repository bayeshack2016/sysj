```python
import pandas as pd
```

```python
df = pd.read_csv('us_counties_pop_april2010_july2015.csv', encoding='UTF-8')
df.head()
```

```python
# POPESTIMATE2015 - 7/1/2015 resident total population estimate
# NPOPCHG2015 - Numeric change in resident total population 7/1/2014 to 7/1/2015
# NATURALINC2015 Natural increase in period 7/1/2014 to 6/30/2015
'''
The key for SUMLEV is as follows:
040 = State and/or Statistical Equivalent
050 = County and /or Statistical Equivalent
The key for REGION is as follows:
1 = Northeast
2 = Midwest
3 = South
4 = West
The key for DIVISION is as follows:
1 = New England
2 = Middle Atlantic
3 = East North Central
4 = West North Central
5 = South Atlantic
6 = East South Central
7 = West South Central
8 = Mountain
9 = Pacific 
'''

# States we'll focus on
ca = df[df.STNAME == 'California']
nv = df[df.STNAME == 'Nevada']
tx = df[df.STNAME == 'Texas']
fl = df[df.STNAME == 'Florida']

ca.head()
```

```python
ca_counties = ca[ca.SUMLEV == 50]
print len(ca)
print len(ca_counties)
```

```python
import pandas_profiling
ca_profile = pandas_profiling.ProfileReport(ca_counties)
ca_profile.to_file(outputfile="ca_pop_profile.html")
```

```python
# just get 2014, 2015 data for ca counties
cols_to_drop = ca_counties.columns
cols_to_drop = [c for c in cols_to_drop if "2014" not in c]
cols_to_drop = [c for c in cols_to_drop if "2015" not in c]
cols_to_drop = [c for c in cols_to_drop if "20" in c]
ca_counties_1415 = ca_counties.drop(cols_to_drop, axis=1)
print ca_counties_1415.head()
print ca_counties_1415.describe()
```
